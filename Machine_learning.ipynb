```python
number=int(input('enter the find its factorial:'))
factorial=1 #initial value is always 1 because  0 factorial is 1 #initiate
if number==0:
  print('the factorial of 0 is 1')

else:
  for i in range(1, number+1):  #range(1,6)->1,2,3,4,5 for the number 5 it means it will include 1 but exclude 6
    factorial =factorial*i #1*1*2
  print('the factorial of ', number, 'is', factorial)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

dataset.head()# give first 5 rows
```

```python
dataset.shape
```

```python
dataset.isnull().sum() # count NaN value
```

## Central Tendencies:

Mean
Median
Mode


```python
# analyse the distribution of data in the salary
fig, ax = plt.subplots(figsize=(8,8))
sns.distplot(dataset.salary)
```

```python
dataset['salary'].fillna(dataset['salary'].median(),inplace=True)#.  inplace = True we are saying that we want the output of our modified data frame to essentially override the original data frame
```

```python

dataset.isnull().sum()
```

```python
#filling missing values with Mean value:
dataset['salary'].fillna(dataset['salary'].mean(),inplace=True)
```

```python
dataset['salary'].fillna(dataset['salary'].mode(),inplace=True)
```
