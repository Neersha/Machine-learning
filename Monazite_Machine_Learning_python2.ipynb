```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
datasets=np.array(iris.data)
datasets
```

```python
covarianceMatrix=pd.DataFrame(data=np.cov(datasets, rowvar=False),columns=iris.feature_names,index=iris.feature_names)
covarianceMatrix
```

```python
eigenvalues, eigenvectors=np.linalg.eig(covarianceMatrix)
eigenvalues
eigenvectors
```

```python
eigenvalues
```

```python
totalSum=sum(eigenvalues)
var_exp=[(i/totalSum)*100 for i in sorted(eigenvalues,reverse=True)]
var_exp
```

```python
featureVector=eigenvectors[:,:2]
featureVector
```

```python
featureVectorTranspose=np.transpose(featureVector)
datasetsTranspose=np.transpose(datasets)
newDatasetsTranspose=np.dot(featureVectorTranspose,datasetsTranspose)
newDatasets=np.transpose(newDatasetsTranspose)
newDatasets
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(19,15))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(newDatasets[:,0],newDatasets[:,1],c=iris.target,s=50)
ax.view_init(30,185)
plt.show()
```

```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
```

```python
iris.get('feature_names')
```

```python
iris.data
```

```python
iris.target
```

```python
pca=PCA(n_components=2)
pca.fit(iris.data)
```

```python
x=pca.transform(iris.data)
x
```

```python
plt.scatter(x[:,0],x[:,1],c=iris.target)
plt.xlabel('First Principal Component-1', fontsize=10)
plt.ylabel('Second Principal Component-2', fontsize=10)
plt.title('2-Dimensional PCA', fontsize=15)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(19,15))
ax=fig.add_subplot(111,projection='3d')
ax.scatter(x[:,0],x[:,1],c=iris.target,s=50)
ax.view_init(30,185)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer=load_breast_cancer()
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
scaled_data=scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(scaled_data)
x_pca=pca.transform(scaled_data)
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma')
```
