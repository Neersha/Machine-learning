```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df2.iloc[:,1:5] = scaler.fit_transform(df2.iloc[:,1:5])
```

```python
import numpy as np
```

```python
covariance_matrix = np.cov([df2.iloc[:,1],df2.iloc[:,2],df2.iloc[:,3],df2.iloc[:,4],df2.iloc[:,5]])
print('Covariance Matrix:\n', covariance_matrix)
```

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
pc = eigen_vectors[0:5]
pc
```

```python
transformed_df2 = np.dot(df2.iloc[:,1:6],pc.T)
```

```python
new_df2 = pd.DataFrame(transformed_df2,columns=['PC1','PC2', 'PC3', 'PC4', 'PC5'])
new_df2['Target'] = df2['Target'].values
new_df2.head()
```
